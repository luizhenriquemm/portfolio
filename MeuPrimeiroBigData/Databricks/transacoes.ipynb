```
import random
import requests
import datetime

url = "http://ec2-52-201-170-42.compute-1.amazonaws.com/recepcao/"

def acao(tipo=None):
  if (tipo == None):
    tipo = random.choice(["deposito", "transferencia", "saque", "saldo", "extrato"])
  
  cliente = random.randint(1, 10000)
  
  if (tipo == "deposito"):
    valor = float("%.2f" % round(random.uniform(1, 1000), 2))
    return {"tipo": tipo, "valor": valor, "cliente": cliente}
    
  elif (tipo == "transferencia"):
    valor = float("%.2f" % round(random.uniform(1, 1000), 2))
    while True:
      cliente_origem = random.randint(1, 10000)
      if (cliente_origem != cliente):
        break
    return {"tipo": tipo, "valor": valor, "cliente": cliente, "cliente-origem": cliente_origem}
  
  elif (tipo == "saque"):
    valor = float("%.2f" % round(random.uniform(1, 1000), 2))
    return {"tipo": tipo, "valor": valor, "cliente": cliente}
  
  elif (tipo == "saldo"):
    return {"tipo": tipo, "cliente": cliente}
  
  elif (tipo == "extrato"):
    return {"tipo": tipo, "cliente": cliente}
  
```

```
print (datetime.datetime.now(), "Iniciando...")

while True:
  inicio = datetime.datetime.now()
  n = 10
  for i in range(n):
    r = requests.post(url, json=acao())

  fim = datetime.datetime.now()
  t = list(map(lambda x: float(x), str(fim - inicio).split(":")))
  t = (t[0] * 60 * 60) + (t[1] * 60) + t[2]
  print ("      Fluxo: " + "{:.2f}".format((n*60)/t) + " transações por minuto", end="\r")
```
